2. Map rows with first row

```python

# Example matrix
data = [
    [2, 4, 6],
    [1, 2, 3],
    [3, 6, 9]
]

first_row = data[0]

result = []

for row in data:
    new_row = []
    for i in range(len(row)):
        new_row.append(row[i] + first_row[i])
    result.append(new_row)

print(result)
```

3. Group similar elements

```python
data = [1, 2, 2, 3, 3, 3, 4]

count = {}

for num in data:
    if num in count:
        count[num] += 1
    else:
        count[num] = 1

print(count)
```

4. Add & subtract matrices

```python
A = [
    [1, 2],
    [3, 4]
]

B = [
    [5, 6],
    [7, 8]
]

add = []
sub = []

for i in range(2):
    add_row = []
    sub_row = []
    for j in range(2):
        add_row.append(A[i][j] + B[i][j])
        sub_row.append(A[i][j] - B[i][j])
    add.append(add_row)
    sub.append(sub_row)

print("Addition:", add)
print("Subtraction:", sub)
```

5. n × n matrix

```python
import random

n = 3

matrix = []

for i in range(n):
    row = []
    for j in range(n):
        row.append(random.randint(1, 9))
    matrix.append(row)

# Make opposite corners even
for i in range(n):
    for j in range(n):
        if (matrix[i][j] + matrix[n-1-i][n-1-j]) % 2 != 0:
            matrix[i][j] += 1

print(matrix)
```

6. Row-wise addition in tuple matrix

```python
matrix = ((1, 2, 3), (4, 5, 6), (7, 8, 9))

result = []

for row in matrix:
    result.append(sum(row))

print(result)
```
